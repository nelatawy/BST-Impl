```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('runtime_data.csv', skipinitialspace=True)

for pct, group in df.groupby('SortedPercent'):
    plt.figure()
    plt.title(f'SortedPercent = {pct}')
    plt.xlabel('Size')
    plt.ylabel('Time (ms)')
    plt.plot(group['Size'], group['Normal Runtime'], label='MergeSort')
    plt.plot(group['Size'], group['RB Runtime'],     label='RB Tree')
    plt.plot(group['Size'], group['BST Runtime'],    label='Simple BST')
    plt.legend()
    plt.show()
```
